```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("diabetesdata.csv")
df.head()
```

```python
inputs=df.drop('result',axis='columns')
target=df['result']
```

```python
inputs
```

```python
target
```

```python
le_FamilyHis=LabelEncoder()
le_Ethni=LabelEncoder()
le_Symptoms=LabelEncoder()
le_Excercise=LabelEncoder()
```

```python
inputs['le_FamilyHis']=le_FamilyHis.fit_transform(inputs['FamilyHis'])
inputs['le_Ethni']=le_Ethni.fit_transform(inputs['Ethni'])
inputs['le_Symptoms']=le_Symptoms.fit_transform(inputs['Symptoms'])
inputs['le_Excercise']=le_Excercise.fit_transform(inputs['Excercise'])
inputs.head()
```

```python

inputs_n=inputs.drop(['FamilyHis','Ethni','Symptoms','Excercise'],axis='columns')
inputs_n
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[45,0,0,0,0]])
```

```python
import pickle
```

```python
with open('diabetic_model','wb') as f:
    pickle.dump(model,f)
```
