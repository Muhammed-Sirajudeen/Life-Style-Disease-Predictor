```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv('strokedata.csv')
```

```python
inputs=data.drop('Result',axis='columns')
target=data['Result']
```

```python
inputs
```

```python
target
```

```python
le_Smoking=LabelEncoder()
le_Alcohol=LabelEncoder()
le_Diabetes=LabelEncoder()
le_BloodPressure=LabelEncoder()
le_Cholestrol=LabelEncoder()
le_Migraine=LabelEncoder()
le_DrugAbuse=LabelEncoder()
le_BirthControlPills=LabelEncoder()
le_Gender=LabelEncoder()
le_Obese=LabelEncoder()
le_PhysicalInactivity=LabelEncoder()
```

```python
inputs['le_Smoking']=le_Smoking.fit_transform(inputs['Smoking'])
inputs['le_Alcohol']=le_Alcohol.fit_transform(inputs['Heavy Drinking'])
inputs['le_Diabetes']=le_Diabetes.fit_transform(inputs['Diabetes'])
inputs['le_BloodPressure']=le_BloodPressure.fit_transform(inputs['BloodPressure'])
inputs['le_Cholestrol']=le_Cholestrol.fit_transform(inputs['Cholestrol'])
inputs['le_Migraine']=le_Migraine.fit_transform(inputs['Migraine'])
inputs['le_DrugAbuse']=le_DrugAbuse.fit_transform(inputs['DrugAbuse'])
inputs['le_BirthControlPills']=le_BirthControlPills.fit_transform(inputs['BirthControlPills'])
inputs['le_Gender']=le_Gender.fit_transform(inputs['Gender'])
inputs['le_Obese']=le_Obese.fit_transform(inputs['Obese'])
inputs['le_PhysicalInactivity']=le_PhysicalInactivity.fit_transform(inputs['Physicalinactivity'])
```

```python
inputs.head()
```

```python
inputs_n=inputs.drop(['Smoking','Heavy Drinking','Diabetes','BloodPressure',
                     'Cholestrol','Migraine','DrugAbuse','BirthControlPills','Gender','Obese','Physicalinactivity'],axis='columns')
inputs_n
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[60,1,1,1,0,0,0,1,0,2,0,1]])
```

```python
import pickle
```

```python
with open('stroke_model','wb') as f:
    pickle.dump(model,f)
```
